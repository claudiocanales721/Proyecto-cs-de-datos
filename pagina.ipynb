# La Maldición de la Segunda Entrega

## 1. Contexto y Motivación

Es común que las franquicias de entretenimiento (cine y videojuegos) lancen secuelas que, a menudo, fracasan rotúndamente, fenómeno que hemos catalogado como "la maldición de la segunda entrega".

Este fracaso recurrente nos lleva a preguntarnos por qué la mayoría de las secuelas no logran replicar el éxito original. La necesidad de un estudio profundo surge ante la falta de evidencia empírica que confirme la percepción popular de que "las segundas partes nunca son buenas".

## 2. Objetivo Principal

Nuestro principal cometido es adentrarnos en este misterio y determinar por qué tantas segundas temporadas o secuelas parecen tropezar donde sus predecesoras triunfaron.

Buscamos evidenciar, con respaldo estadístico y cualitativo, a qué se debe la recurrencia de estos fracasos, transformando una percepción común en una conclusión investigativa sólida que aporte claridad a las empresas que piensen en expandir sus franquicias.

## 3. Preguntas de Investigación

¿El tiempo transcurrido entre el lanzamiento del juego original y su secuela influye en la diferencia de puntuación? Específicamente, ¿las secuelas lanzadas más rápidamente tienden a tener peores calificaciones?

¿La puntuación de los usuarios de las secuelas es más baja que la de los juegos originales, y es esta caída mayor o menor que la de la crítica profesional?

¿El cambio de desarrollador entre la entrega original y la secuela se correlaciona con una mayor caída en las puntuaciones?

¿Los factores asociados al fracaso son manejables para el desarrollador?

## 4. Diseño Tentativo y Metodología

Nuestro objetivo es encontrar una relación cuantificable entre las distintas variables de producción de un videojuego (como el tiempo de desarrollo, el equipo creativo o el género) y su recepción final.

Para esto, usaremos Python como herramienta principal, aplicando conceptos de estadística básica y librerías clave:

**Pandas** y **Numpy**: Para el manejo y procesamiento de la información.

**Geopandas** y **Matplot**: Para la visualización y análisis de datos.

### Datos

Exploraremos bases de datos de Kaggle y APIs con columnas como nombre, ID, fecha de estreno, género, rating y valoraciones de usuarios (IMDb, Rotten Tomatoes). Aplicaremos técnicas de limpieza como `dropna`, `drop_duplicates` y `fillna`.
