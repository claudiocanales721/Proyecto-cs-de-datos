```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
import seaborn as sns
df_movies = pd.read_csv(os.path.join("data_claudio", "movies_dataset.csv"))
"""Primero que nada haremos un diagrama de caja para ver la distribucion de las calificaciones dadas por IMDb y RottenTomatoes
Empezaremos conn IMDb (Considerando su escala de calificacion del 1 al 10)"""

plt.figure(figsize=(6, 8)) 
sns.boxplot(y=df_movies["IMDbRating"], color="red")
plt.title('Distribuci贸n de Calificaciones IMDb')
plt.ylabel('Calificaci贸n (1-10)')
plt.show()

"""Ahora haremos lo mismo pero con Rotten Tomatoes(Considerando un rating de 1 a 100)"""
plt.figure(figsize=(6, 8)) 
sns.boxplot(y=df_movies["RottenTomatoesScore"], color="green")
plt.title('Distribuci贸n de Calificaciones Rotten Tomatoes')
plt.ylabel('Calificaci贸n (1-100)')
plt.show()
```

```python
"""Observamos que la distribucion de calificaciones para ambas plataformas es muy similar"""
```

```python
valores_por_paises = []
for element in df_movies["Country"].unique():
    valores_por_paises.append(df_movies["Country"].value_counts()[element])
"""Ahora con las cantidades de peliculas por pais haremos un grafico de barras para comparar la produccion por pais"""
etiquetas = [x for x in df_movies["Country"].unique()]
fig, ax = plt.subplots(figsize=(15, 6))

ax.bar(etiquetas, valores_por_paises)

ax.set_title("Pais")
ax.set_ylabel("Cantidad de peliculas")

plt.show()
```

```python
"""Observamos una diferencia abismal entre la cantidad de peliculas producidas por Estados Unidos frente a las demas"""
```

```python
"""Ahora haremos exactamente lo mismo pero ahora con la cantidad de generos por pelicula"""
valores_por_genero = []
for element in df_movies["Genre"].unique():
    valores_por_genero.append(df_movies["Genre"].value_counts()[element])
"""Del mismo modo, haremos un grafico para observar mejor la distribucion de genero por pelicula"""
etiquetas = [x for x in df_movies["Genre"].unique()]
fig, ax = plt.subplots(figsize=(15, 6))

ax.barh(etiquetas,valores_por_genero)

ax.set_title("Genero")
ax.set_ylabel("Cantidad de peliculas con el genero")

plt.show()
```

```python
"""De aca observamos que los generos mas producidos son el horror, la comedia y la accion"""
```

```python
"""Ahora me dispongo a ver la correlacion entre los ingresos generados por la venta de entradas en USD y la valoracion. Primero en IMDb y luego en Rotten Tomatoes"""
fig, ax = plt.subplots(figsize=(10, 6))

sns.regplot(data=df_movies, x="US_BoxOfficeUSD", y="IMDbRating", ax=ax, scatter_kws={'alpha': 0.2}, line_kws={'color': 'red'})
ax.set_xscale('log')
ax.set_title("Relaci贸n entre ingresos en USD y Valoraci贸n (IMDb)")
ax.set_xlabel("Ingresos generados por la venta de entradas en USD (Escala logaritmica)")
ax.set_ylabel("Valoraci贸n en IMDb")
plt.show()
```

```python
"""Lo mismo ahora con Rotten Tomatttoes"""
fig, ax = plt.subplots(figsize=(10, 6))

sns.regplot(data=df_movies, x="US_BoxOfficeUSD", y="RottenTomatoesScore", ax=ax, scatter_kws={'alpha': 0.2}, line_kws={'color': 'red'})
ax.set_xscale('log')
ax.set_title("Relaci贸n entre ingresos en USD y Valoraci贸n (Rotten Tomatoes)")
ax.set_xlabel("Ingresos generados por la venta de entradas en USD (Escala logaritmica)")
ax.set_ylabel("Valoraci贸n en IMDb")
plt.show()
```

```python
"""Se observa un correlacion practicamente nula debido a que la linea de tendencia es casi horizontal (pendiente 0) por lo que no se puede
establecer una relacion directa entre los ingresos generados y la calificacion obtenida en Rotten Tomatoes e IMDb.
Es por esto que se concluye que saber la cantidad de ingresos que genero una pelicula no te proporciona informacion acerca de su calificacion y viseversa"""
```
